```python
# Select numeric columns (corrected line)
dataset_numeric = df.select_dtypes(include=['number'])  # No backslash or extra characters

# Compute the correlation matrix
cor_matrix = dataset_numeric.corr()

# Heatmap part
plt.figure(figsize=(10, 8))
sns.heatmap(cor_matrix,
            annot=True,
            cmap="coolwarm",
            fmt=".2f",
            linewidths=0.5,
            square=True,
            cbar_kws={'label': 'Correlation'})

# Customize the plot
plt.title("Correlation Matrix of Financial Data", fontsize=16, pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()

# Display the heatmap
plt.show()
```
